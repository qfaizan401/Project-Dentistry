```python
import os
import numpy as np
import pandas as pd
```

```python
from skimage.measure import regionprops_table
from skimage import io, measure, img_as_ubyte
from skimage.filters import threshold_multiotsu
from skimage.color import label2rgb
```

```python
img_dir = 'catagorical_data/4-root_canal_treated/'
img_list = os.listdir(img_dir)
len(img_list)
```

```python
properties = ['label', 'area', 'bbox', 'bbox_area', 'centroid', 'convex_area', 'eccentricity', 'equivalent_diameter', 
              'euler_number', 'extent', 'filled_area', 'inertia_tensor_eigvals', 'local_centroid', 'major_axis_length', 
              'minor_axis_length', 'moments_hu', 'orientation', 'perimeter', 'slice', 'solidity']
csv_dir = 'Final_CSVs/root_canal_treated/'

for i in range(len(img_list)):
    img = io.imread(os.path.join(img_dir, img_list[i]), as_gray=True)
    threshold = threshold_multiotsu(img)
    region = img_as_ubyte(np.digitize(img, bins=threshold))
    label_img = measure.label(region, connectivity=img.ndim)
    
    img_props = measure.regionprops_table(label_img, img, properties=properties) 
    temp_img = pd.DataFrame(img_props)
    temp_img = temp_img[temp_img['area']>70]
    temp_img.to_csv(os.path.join(csv_dir, f'{img_list[i]}.csv'))
```
